```python
def estimate_time(p, q, bit_time, mul_time):
    found = False
    max_NK = 0
    best_combo = (p, q, 0, 0, 0)
    for N in range(1, 10):
        for K in range(1, 10):
            for G in range(6):
                S = p + q + G
                if p+(N-1)*S <= A_bits and q+(K-1)*S <= W_bits and G >= math.log(min(N, K), 2):
                    if N*K > max_NK:
                        best_combo = (p, q, N, K, G)
                        max_NK = N*K
    N = best_combo[2]
    K = best_combo[3]
    print('p = %d, q = %d, N = %d, K = %d, G = %d'%(best_combo[0], best_combo[1], best_combo[2], best_combo[3], best_combo[4]))
    theo_re = N*K
    print('Each loop takes %d multiplications, %d shifts, and %d and/or gates'%(1, ((N - 1)*2 + 2), ((N - 1)*2 + 1)))
    expe_re = (N*K*mul_time)/(((N - 1)*2 + 2)*bit_time + mul_time)
    print('Theoretical  reduction = ', theo_re)
    print('Experimental reduction = ', expe_re)
    print('===============================================')
```

```python
bit_time = 1.1
mul_time = 1.4
for p in range(1, 8):
    for q in range(1, p + 1):
        estimate_time(p, q, bit_time, mul_time)
```
